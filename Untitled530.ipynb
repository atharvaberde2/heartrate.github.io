```python
import pandas as pd
```

```python
df = pd.read_csv('heart_statlog_cleveland_hungary_final.csv')
```

```python
X = df.drop(columns = 'target')
```

```python
Y= df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Initialize Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_clf.fit(X_train, y_train)

# Make predictions
y_pred = rf_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Feature Importance (Optional)
feature_importances = pd.DataFrame(rf_clf.feature_importances_,
                                   index=X.columns,
                                   columns=['Importance']).sort_values('Importance', ascending=False)
print("\nFeature Importances:")
print(feature_importances)
```

```python
import joblib
import pickle
joblib.dump(rf_clf, 'heartrate_classification_model.pkl')

# or Save the trained model to a .pkl file using pickle
with open('C:/Users/atber/Downloads/heartrate_classification_model.pkl', 'wb') as f:
    pickle.dump(rf_clf, f)
```
